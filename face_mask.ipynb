```python
### DATA-agumentation dataframe

from keras.preprocessing.image import ImageDataGenerator
import cv2

```

```python
train_ge=ImageDataGenerator(rescale=1./255,rotation_range=90,height_shift_range=0.2,brightness_range=[0,1],shear_range=0.1)

```

```python
valid_ge=ImageDataGenerator(rescale=1./255)

```

```python
train_de=train_ge.flow_from_directory(r"C:\Users\Irfan\Deep_Learning\Assignments\f_mask\train",target_size=(200,200),batch_size=10,class_mode='categorical')

```

```python
valid_de=valid_ge.flow_from_directory(r"C:\Users\Irfan\Deep_Learning\Assignments\f_mask\validation",target_size=(200,200),batch_size=10,class_mode='categorical')
```

```python
def tuner(par):
    model = Sequential()
    x = 0
    for i in range(par.Int('Conv-Layers',2,30)):
        model.add(Conv2D(par.Int("filters{}".format(x),5,10),(5,5),activation=par.Choice('Activation',values=['sigmoid','relu','tanh']),input_shape = (200,200,3)))
        x+=1
    model.add(AveragePooling2D(pool_size=(2,2), strides=2))
    model.add(Flatten())
    x = 0
    for i in range(par.Int('Hidden-Layers',2,30)):
        model.add(Dense(par.Int("neurons{}".format(x),5,20),activation=par.Choice('Activation',values=['sigmoid','relu','tanh'])))
        
        x+=1
    model.add(Dense(2,activation = 'softmax'))
    model.compile(optimizer=par.Choice('opti',values=['rmsprop','Adam']),loss ='categorical_crossentropy',metrics=['accuracy'] )
    return model
```

```python
robj=keras_tuner.RandomSearch(tuner,objective="val_accuracy",max_trials=10,directory=r"C:\Users\Irfan\Desktop\Face_mask_")
```

```python
robj.search(train_de,epochs=5,validation_data=valid_de)
```
